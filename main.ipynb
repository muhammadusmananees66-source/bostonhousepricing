**Preparing Dataset**

```python
dataset = pd.DataFrame(boston.data, columns = boston.feature_names)
```

```python
# Create a Model 
# Independent and dependent features 
x = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
# Train Test Split 
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = .3, random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
# Standardize all teh data at the same scale 
# Standardize the dataset
from sklearn.preprocessing import StandardScaler
Scaler = StandardScaler()
```

```python
x_train = Scaler.fit_transform(x_train)
x_test = Scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

**Model Training**

```python
from sklearn.linear_model import LinearRegression 
regression = LinearRegression()
regression.fit(x_train, y_train)
```

```python
# In case fo regression Modle we need to take into account two factors 
# coeffcicent and intercept 
print(regression.coef_) # This would be equal to all number of input features(means for each independent feature we have coefficient)
 
```

```python
print(regression.intercept_)
```

```python
# On which parameters the model has been trained
regression.get_params()
```

**Prediction of the Model**

```python
# Prediction of the model with test data 
reg_pred = regression.predict(x_test)
```

```python
reg_pred
```

**Assumptions**

```python
# Plot some diagram to confirm that the predictions matches to the actual values
# Plot a scatter plot for the prediction 
plt.scatter(y_test, reg_pred)
# Whe the reg_pred and y_test go in linear so it means that the prediction is right 
# Here are some points that are away from the graph so these are outliers 
```

```python
# Now calculate residual means error 
residuals= y_test - reg_pred
```

```python
residuals
```

```python
# Plotting the residuals 
# As from the graph there are some outliers at the right hand side
sns.displot(residuals, kind='kde' )
```

```python
# Scatter plot with respect to prediction and residuals 
# Here is a uniform distribution
# so with all above graph above this give us suggestions that the model is performing well 
plt.scatter(reg_pred, residuals)
```

```python
# To be sure how model is performing we have to use some kind of metrics
from sklearn.metrics import mean_squared_error 
from sklearn.metrics import mean_absolute_error 

print(mean_absolute_error(y_test, reg_pred))
print(mean_squared_error(y_test, reg_pred))
print(np.sqrt(mean_squared_error(y_test, reg_pred)))

```

```python
# R Square and adjusted R squre 
# With more metrics we are going to check that the linear regression model is performing well or not 
# Always Adjusted R square woudl be less than the R Square
from sklearn.metrics import r2_score
score = r2_score(y_test, reg_pred)
```

```python
# As its .71 so as mroe its near to 1 its better to get more score 
print(score)
```

```python
# for adjusted r squre there is no any library so we will calculate it as 
1 - (1-score)*(len(y_test)-1)/(len(y_test)-x_test.shape[1]-1)
```
